```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
from sklearn.linear_model import LinearRegression
from sklearn.
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
import plotly.express as px
from sklearn.compose import TransformedTargetRegressor
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/House%20Prices.csv",index_col="ID")
df.columns
```

```python
   print(ols(f"Price~Date",data=df).fit().summary())
```

```python
x=df[['Bedrooms', 'Bathrooms', 'Sqft_living',
       'Sqft_lot', 'Floors', 'Waterfront', 'View', 'Condition', 'Grade',
       'Sqft_above', 'Sqft_basement', 'Yr_built', 'Yr_renovated', 'zipcode',
       'Lat',  'Sqft_living15', 'Sqft_lot15']]
y=df["Price"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.25,random_state=0)
```

```python
l1=LinearRegression()
l1.fit(xtrain,ytrain)
```

```python
y_pre=l1.predict(xtest)
```

```python
r2_score(y_true=ytest,y_pred=y_pre)
```
